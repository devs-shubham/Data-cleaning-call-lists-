```python
import pandas as pd
df = pd.read_excel("/content/Customer Call List.xlsx")
print(df)
```

```python
df.head(5)
```

```python
df =df.drop_duplicates()
df
```

```python
df = df.drop(columns="Not_Useful_Column")
df
```

```python
# df["Last_Name"] = df["Last_Name"] .str.lstrip("...")
# df["Last_Name"] = df["Last_Name"] .str.rstrip("_")
# df["Last_Name"] = df["Last_Name"] .str.lstrip("/")

df["Last_Name"] = df["Last_Name"] .str.strip("123._/")
df
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].astype(str).str.replace(r"\D", "", regex=True)
df
```

FORMATTING

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:9])
df

```

```python

df["Address"] = df["Address"].str.replace('N/a' , '')
df
```

Split a column into multiple columns (most common)


```python
df[["street_Address" , "state", "zip_code"]]= df["Address"].str.split(pat=',' ,n=2, expand=True)
df
```

```python
df = df.drop(columns="Address")
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('N/a' , '')
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes' , 'Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No' , 'N')
df
```

```python

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes' , 'Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No' , 'N')
df
```

```python
df = df.fillna('')
df
```

```python
df
```

DELETING row where do not contact = Y

```python
for x in df. index:
  if df.loc[x, "Do_Not_Contact"] == "Y":
    df.drop(x, inplace=True)
df
```

deleting empty phone no row


```python
for x in df. index:
  if df.loc[x, "Phone_Number"] == "--":
    df.drop(x, inplace=True)
df

# another way to drop
# df = df.dropna(subset = "Phone_Number"), inplace = true)
```
